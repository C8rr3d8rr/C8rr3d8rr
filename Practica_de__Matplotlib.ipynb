```python
#Colores basicos para las lineas de los graficos

#r = red
#b = blue
#g = green

# -. linea con puntos
# -- linea descontinuada
plt.plot(x, x**2, 'b.-') # linea azul con puntos
plt.plot(x, x**3, 'g--') # Linea verde discontinua
plt.plot(x, x**1, 'r-.') # Linea roja con puntos
```
