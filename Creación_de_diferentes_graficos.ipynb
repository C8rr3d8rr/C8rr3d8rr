```python
#Importacion de las librerias pandas y matplotlib
import pandas            as pd
import matplotlib.pyplot as plt
import numpy             as np
```

```python
#Se sube el data set
df=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Pruebas/pruebas.xlsx')
```

```python
#Se verifica la informacion del data set
df.head()
```

```python
#asignacion de variables a los ejes  x & y

x=df.Unidades_vendidas
y=df.Edad

plt.scatter(x,y) #asignacion de variables
plt.title("Scatter plot Simple") #asignacion del titulo
plt.xlabel('Unidades_vendidas') #asignacion del titulo del eje x
plt.ylabel('Edad de los vendedores') #asignacion de titulo del eje y
plt.grid(True) #creacion de una reticula
```

```python
#creacion de grafico de cajas y bigotes (Boxplot)

#crear datos aleatorios
data = [np.random.normal(0, std, 200) for std in range(1, 6)]

# boxplot rectangular
plt.boxplot(data,vert=True,patch_artist=True);
plt.title("Cajas y bigotes")
```

```python
#Diagrama de torta

labels = 'Colombia', 'Mexico', 'Peru', 'Venezuela'
sizes = [20, 50, 80, 5]
explode = (0, 0.1, 0, 0)  # solo "Saque" el 2do pedazo (ejem. 'cerdos')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis('equal')  #Homogeniadidad del grafico
plt.show()
```
